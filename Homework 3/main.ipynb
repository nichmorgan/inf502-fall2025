# Homework 3 - Divide and Conquer

A divide and conquer GCD requires a recursive function. The recursion stops when the second parameter (`b`) equals 0. At each recursive step, we replace `(a, b)` with `(b, a % b)`, where `a % b` is the remainder of dividing `a` by `b`.

## Example: GCD(48, 30)

- GCD(48, 30) → 30 != 0, calculate GCD(30, 48%30) = GCD(30, 18)
- GCD(30, 18) → 18 != 0, calculate GCD(18, 30%18) = GCD(18, 12)
- GCD(18, 12) → 12 != 0, calculate GCD(12, 18%12) = GCD(12, 6)
- GCD(12, 6) → 6 != 0, calculate GCD(6, 12%6) = GCD(6, 0)
- GCD(6, 0) → 0 = 0, **return 6**

```python
def GCD(a: int, b: int) -> int:
    if b == 0:
        return a
    return GCD(b, a % b)

GCD(48, 30)
```
